```python
def convert_size(size):
  if 'M' in size:
    return int(float(size.replace('M',''))*1000)
  elif 'K' in size:
    return float(size.replace('k',''))
  elif 'Varies with device' in size:
    return np.nan
  else:
    return np.nan
```

```python
df['Size']=df['Size'].apply(convert_size)
```

```python
cols=['Installs','Price']
chars_to_remove=['+',',','$']
for i in cols:
  for j in chars_to_remove:
    df[i]=df[i].str.replace(j,'')
```

```python
#Q14  What are the top 10 most popular apps in the dataset?
# Create a frequency table to identify the apps with the highest number of installs.
app_installs=df.groupby('App')['Installs'].sum()
top_10_apps=app_installs.sort_values(ascending=False).head(10)
frequency_table = df[df['App'].isin(top_10_apps.index)]['App'].value_counts()

# Print the frequency table
print("Frequency table for the top 10 most popular apps:")
print(frequency_table)
colors=['red','blue','green','orange','teal']
top_10_apps.plot(kind='bar',x='Installs',y='App',color=colors)
```

```python
#15A company wants to launch a new app on the Google Playstore and has asked you to analyze the Google Playstore dataset
# to identify the most popular app categories. How would you approach this task, and what features would you analyze to make
#  recommendations to the company?
category_rating=df.groupby('Category')['Rating'].mean()
top_10_category_rating=category_rating.sort_values(ascending=False).head(10)

top_10_category_rating.plot(kind='bar',x='Category',y='Rating')
```

```python
category_installs=df.groupby('Category')['Installs'].sum()
top_10_category=category_installs.sort_values(ascending=False).head(10)

top_10_category.plot(kind='bar',x='Category',y='Installs')
```

```python
#Q16. A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers.
# What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?
df_merged = pd.DataFrame({'Installs': top_10_category, 'Rating': top_10_category_rating})

# Plot the combined DataFrame
plt.ylabel('Installs')
plt.xlabel('Category')
plt.figure(figsize=(20,10))
plt.title('Top Categories by Installs and Ratings')
df_merged.plot(kind='bar', secondary_y='Rating',rot=45)
```

```python
#Q17. A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to launch a new app.
# What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?
# Convert the Last Update into datetime datatype
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
monthUpdate = df['Last Updated'].dt.month
installPerMonth = df.groupby(monthUpdate)['Installs'].sum()
print(installPerMonth)
sns.barplot(x=installPerMonth.index, y=installPerMonth.values, palette='pastel')
plt.xlabel('Last Updated per Month')
plt.ylabel('Installs per Month')
plt.show()
```
